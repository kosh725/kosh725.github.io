```python
# API initialisation
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
#Data Storage: skeleton data frame
#Lockdown
lock_tweet = []
lock_like = []
lock_time = []
lock_current_month = []

#Vaccines
vac_tweet = []
vac_like = []
vac_time = []
vac_current_month = []

#Job Keeper
job_tweet = []
job_like = []
job_time = []
job_current_month = []

#Working from home
wfh_tweet = []
wfh_like = []
wfh_time = []
wfh_current_month = []

#Mortality
mort_tweet = []
mort_like = []
mort_time = []
mort_current_month = []
```

```python
#Cursor intialisation to browse tweets
tweet_number = 33000
lockdown = tweepy.Cursor(api.search, q="lockdown", tweet_mode="extended").items(tweet_number)
vaccine = tweepy.Cursor(api.search, q="vaccine", tweet_mode="extended").items(tweet_number)
job_keeper = tweepy.Cursor(api.search, q="jobkeeper", tweet_mode="extended").items(tweet_number)
working_from_home = tweepy.Cursor(api.search, q="wfh", tweet_mode="extended").items(tweet_number)
mortality = tweepy.Cursor(api.search, q="death rate", tweet_mode="extended").items(tweet_number)
```

```python
#run to find the attributes searchable in each tweet
attribute = tweepy.Cursor(api.search,tweet_mode="extended").items(1)
for x in attribute:
    print(dir(x))
```

```python
#extracting lockdown data from the cursor
for x in lockdown:
    lock_tweet.append(x.full_text)
    lock_like.append(x.favorite_count)
    lock_time.append(x.created_at)
#creating lockdown dataframe
df_lck = pd.DataFrame({'tweet':lock_tweet, 'like':lock_like, 'time': lock_time})
filt = (df_lck['like'] > 0)

#df_lck[filt] -> filtered array of data

#count the tweets and push to the total
num_tweet = len(df_lck[filt].index)
lock_current_month.append(num_tweet)
```

```python
#extracting vaccine data from the cursor
for x in vaccine:
    vac_tweet.append(x.full_text)
    vac_like.append(x.favorite_count)
    vac_time.append(x.created_at)

df_vac = pd.DataFrame({'tweet':vac_tweet, 'like':vac_like, 'time': vac_time})
filt = (df_vac['like'] > 0)

num_vaccine_tweet = len(df_vac[filt].index)
vac_current_month.append(num_vaccine_tweet)
```

```python
#extracting job keeper data from the cursor
for x in job_keeper:
    job_tweet.append(x.full_text)
    job_like.append(x.favorite_count)
    job_time.append(x.created_at)

df_job = pd.DataFrame({'tweet':job_tweet, 'like':job_like, 'time': job_time})
filt = (df_job['like'] > 0)

num_job_keeper_tweet = len(df_job[filt].index)
job_current_month.append(num_job_keeper_tweet)
```

```python
#extracting working from home data from the cursor
for x in working_from_home:
    wfh_tweet.append(x.full_text)
    wfh_like.append(x.favorite_count)
    wfh_time.append(x.created_at)

df_wfh = pd.DataFrame({'tweet':wfh_tweet, 'like':wfh_like, 'time': wfh_time})
filt = (df_wfh['like'] > 0)

num_working_from_home_tweet = len(df_wfh[filt].index)
wfh_current_month.append(num_job_keeper_tweet)
```

```python
#extracting working from home data from the cursor
for x in mortality:
    mort_tweet.append(x.full_text)
    mort_like.append(x.favorite_count)
    mort_time.append(x.created_at)

df_mort = pd.DataFrame({'tweet':mort_tweet, 'like':mort_like, 'time': mort_time})

filt = (df_mort['like'] > 0)
num_mortality_tweet = len(df_mort[filt].index)
mort_current_month.append(num_mortality_tweet)
```

```python
#Create the final monthly tally of tweets data frame
df_month = pd.DataFrame({'Lockdown':lock_current_month, 'Vaccine':vac_current_month,'JobKeeper':job_current_month,
                        'Working from home':wfh_current_month, 'Mortality':mort_current_month})
```
